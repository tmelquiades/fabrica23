# REVISAR 
Quantidade de acidentes por estado:

```python
dfdrug = dfLimpo.groupby('Drug Involvement')[['Drug Involvement']].count()
```

```python
dfdrug
```

Índice de envolvimento com drogas das pessoas envolvidas nos acidentes

```python
dfdrug.plot(kind = 'pie', subplots = True)
```

Obversa-se que a maioria não foi reportada, mais de 1900 foram negativos e apenas 200 foram positivos. 

Quantidade de acidentes fatais por estado:

```python
dfFatal = dfLimpo.groupby('State')[['Fatalities in crash']].count().plot(kind='barh', figsize=(10,8), color = 'b')
```

Observa-se que o estado com maior quantidade de acidentes fatais foi Wyoming, com 129. O estado com menor quantidade foi District of Columbia, com 48.
