# Agrupar o BD a partir da raça/etnia dos alunos
e plotar
um gráfico com a nota média em Matemática para cada grupo
disponível

```python
## criando outro dataframe pra armazenar esse agrupamento
df_mathmean = df.groupby('race/ethnicity')[['math score']].mean()
df_mathmean ## conferir a tabela...
```

```python
df_mathmean.plot(kind='bar', color='blue') ## plotar o gráfico
```

Concluí-se que os alunos do grupo E obtiveram a maior média de notas em matemática dentre o agrupamento por etnias, com uma média de 73,8, e o grupo A obteve a menor média, de 61,6.

# Agrupar o BD a partir do Nível de Escolaridade dos pais
e plotar um gráfico com a nota média em Escrita para cada grupo

```python
## criando outro df pra armazenar esse agrupamento
df_writingmean = df.groupby('parental level of education')[['writing score']].mean()
df_writingmean ## conferindo a tabela
```

```python
df_writingmean.plot(kind='bar', color='green') ## plotando o gráfico
```

Alunos cujos pais cursaram apenas o ensino médio obtiveram a menor média na disciplina de escrita, de 62,4, enquanto os alunos cujos pais fizeram mestrado obtiveram a maior média, de 75,6.

# Agrupar o banco de dados a partir da coluna referente à partificipação
em um curso preparatório para os testes, e plotar um gráfico com a
nota média em Leitura para cada grupo disponível.

```python
## agrupando...
df_readingscore = df.groupby('test preparation course')[['reading score']].mean()
df_readingscore
```

```python
## plotando o grafico

df_readingscore.plot(kind='pie', subplots=True, figsize=(6,6))
```

A média dos alunos que completaram o curso de preparamento para a prova é de 7,3 pontos maior que a média dos que não completaram o curso.
