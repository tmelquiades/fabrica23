# Agrupamento de dados e plotagem de gr√°ficos

# Plote um gr√°fico de barras (bar) que exiba a quantidade de funcion√°rios em cada setor (coluna Department)

```python
df_qfd = df.groupby('Department')[['EmployeeCount']].count()
df_qfd ##Q = quantidade, F = funcion√°rios, D = departamento
```

```python
df_qfd.plot(kind='bar', color='purple')
```

Conclui-se que a maioria dos funcion√°rios dessa empresa s√£o do departamento de Research and Development (Pesquisa e Desenvolvimento), com 961 funcion√°rios.<br>O departamento com menos funcion√°rios √© o de Human Resources (Recursos Humanos), com 63.<br>E o de Sales (Vendas) tem 446 funcion√°rios.

# Plote um gr√°fico de linha (line) que exiba a quantidade de funcion√°rios para cada n√≠vel de escolaridade (coluna Education)

```python
df_qfne = df.groupby('Education')[['EmployeeCount']].count()
df_qfne ## Q = quantidade, F = funcion√°rios, NE = n√≠vel de escolaridade
```

```python
df_qfne.plot(kind='line', color='orange')
```

Observa-se que 572 funcion√°rios t√™m at√© o terceiro n√≠vel de escolaridade e apenas 48 t√™m at√© o n√≠vel 5.<br>H√° mais funcion√°rios com at√© o quarto n√≠vel de escolaridade do que com apenas segundo ou primeiro.

# Plote um gr√°fico de torta/pizza (pie) que exiba a quantidade de funcion√°rios que tem ou n√£o algum atrito na empresa (coluna Attrition)

```python
df_qfa = df.groupby('Attrition')[['EmployeeCount']].count()
df_qfa ## Q = quantidade, F = funcion√°rios, A = atrito
```

```python
df_qfa.plot(kind='pie', subplots=True)
```

Observa-se que a maioria dos funcion√°rios n√£o t√™m atrito na empresa, apenas 237 dos 1470 t√™m algum atrito.
