# Agrupamento de dados e plotagem de gráficos

# Plote um gráfico de barras (bar) que exiba a quantidade de funcionários em cada setor (coluna Department)

```python
df_qfd = df.groupby('Department')[['EmployeeCount']].count()
df_qfd ##Q = quantidade, F = funcionários, D = departamento
```

```python
df_qfd.plot(kind='bar', color='purple')
```

Conclui-se que a maioria dos funcionários dessa empresa são do departamento de Research and Development (Pesquisa e Desenvolvimento), com 961 funcionários.<br>O departamento com menos funcionários é o de Human Resources (Recursos Humanos), com 63.<br>E o de Sales (Vendas) tem 446 funcionários.

# Plote um gráfico de linha (line) que exiba a quantidade de funcionários para cada nível de escolaridade (coluna Education)

```python
df_qfne = df.groupby('Education')[['EmployeeCount']].count()
df_qfne ## Q = quantidade, F = funcionários, NE = nível de escolaridade
```

```python
df_qfne.plot(kind='line', color='orange')
```

Observa-se que 572 funcionários têm até o terceiro nível de escolaridade e apenas 48 têm até o nível 5.<br>Há mais funcionários com até o quarto nível de escolaridade do que com apenas segundo ou primeiro.

# Plote um gráfico de torta/pizza (pie) que exiba a quantidade de funcionários que tem ou não algum atrito na empresa (coluna Attrition)

```python
df_qfa = df.groupby('Attrition')[['EmployeeCount']].count()
df_qfa ## Q = quantidade, F = funcionários, A = atrito
```

```python
df_qfa.plot(kind='pie', subplots=True)
```

Observa-se que a maioria dos funcionários não têm atrito na empresa, apenas 237 dos 1470 têm algum atrito.
